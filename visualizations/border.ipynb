```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, '--r')
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, '--r')
plt.xlim(-10, 20)
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, 'r')
plt.xlim(-1, 11)
plt.ylim(-10, 120)
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, 'r')
# plt.xlim(-1, 11)
# plt.ylim(-10, 120)
plt.axis([-1, 11, -10, 110])
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, 'r')

plt.axis('tight')
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y, 'r')

plt.axis('equal')
```
