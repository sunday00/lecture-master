```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
```

```python
plt.plot(x, np.cos(x), '--')
```

```python
plt.plot(x, np.tan(x), '--^')
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.plot(x, np.tan(x), '--^')
```

```python
fig.savefig('test.png')
```
