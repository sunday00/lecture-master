```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = x ** 2
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r')
plt.xlabel('this is X')
plt.ylabel('this is Y')
plt.show()
```

```python
y1 = x ** 2
y2 = x ** 3
plt.plot(x, y1, 'r', label="line1")
plt.plot(x, y2, 'g', label="line2")
plt.xlabel('this is X')
plt.ylabel('this is Y')

plt.legend()

plt.show()
```
