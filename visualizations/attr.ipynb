```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 20)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y, linewidth=3, color="#ffff00", marker="^", markersize=10, markerfacecolor='green')
plt.show()
```

```python
# plt.plot(x, y, linewidth=3, color="#ffff00", marker="^", markersize=10, markerfacecolor='green')
plt.plot(x, y, 'k:', linewidth=3, marker="^", markersize=10, markerfacecolor='green')
plt.show()
```
