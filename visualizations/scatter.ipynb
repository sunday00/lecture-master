```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.scatter(20,20,s=100, c='r', alpha=0.5)
plt.show()
```

```python
plt.scatter([10, 20],[10, 20], s=100, c=['r', 'g'], alpha=0.5)
plt.show()
```

```python
x = np.linspace(0, 10, 20)
y = x ** 2

plt.scatter(x, y, s=100, c='r', alpha=0.5)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
c = np.random.rand(50)
s = np.random.rand(50) * 500

plt.scatter(x,y,s=s, c=c, alpha=0.5)
plt.show()
```
