```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

labels = ['one', 'two', 'three', 'four'] 
sizes = [10, 20, 30, 40]
```

```python
plt.pie(sizes, labels=labels)
plt.show()
```

```python
plt.pie(sizes, labels=labels, shadow=True, startangle=90, counterclock=False)
plt.show()
```

```python
sizes = [10, 20, 30, 40]
explode = (0,0,0.1,0)

plt.pie(sizes, labels=labels, shadow=True, startangle=90, counterclock=False, explode=explode)

plt.show()
```

```python
sizes = [10, 20, 30, 40]
explode = (0,0,0.1,0)

plt.pie(sizes, labels=labels, shadow=True, startangle=90, counterclock=False, explode=explode, autopct="% 1.1f %%")

plt.show()
```

```python
plt.bar(labels, sizes)
plt.show()
```

```python
plt.bar(labels, sizes)
plt.title('Sales count', fontsize=17)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.show()
```

```python
plt.barh(labels, sizes)
plt.title('Sales count', fontsize=17)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.show()
```

```python
print( sizes )

plt.barh(labels, sizes, color="g")
plt.barh(labels, [20, 40, 60, 80], color="b", left=sizes)
# plt.bar(labels, [20, 40, 60, 80], color="b", bottom=sizes)
plt.title('Sales count', fontsize=17)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.show()
```

```python
print( sizes )

plt.barh(labels, sizes, color="g", label="last")
plt.barh(labels, [20, 40, 60, 80], color="b", left=sizes, label="current")

plt.title('Sales count', fontsize=17)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Sales', fontsize=15)

plt.legend()

plt.show()
```

```python
print( sizes )

plt.barh(labels, sizes, color="g")
plt.barh(labels, [20, 40, 60, 80], color="b", left=sizes)

plt.title('Sales count', fontsize=17)
plt.xlabel('Items', fontsize=15)
plt.ylabel('Sales', fontsize=15)

plt.legend(['last', 'current'])

plt.show()
```
