```python
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y)

plt.show()
```

```python
y_ = x ** 3

plt.plot(x, y)
plt.plot(x, y_)
```

```python
plt.plot(x, y, color="crimson")
plt.plot(x, y_, color="#456DF3")

ax = plt.gca()
ax.set_facecolor('k')
plt.show()
```

```python
fig = plt.figure()
fig.patch.set_facecolor('xkcd:mint green')

plt.plot(x, y, color="crimson")
plt.plot(x, y_, color="#1d59c9")

ax = plt.gca()
ax.set_facecolor('k')

plt.show()
```
