```python
import numpy as np
import pandas as pd
import random
import time

random.seed(time.time())

careers = [int(i) for i in range(1, 7)]
years = [i for i in range(2015, 2021)]
sails = []
for i in range(1, 7) :
  sails.append( random.randint(1000000, 20000000) )

profit = []
for s in sails :
  profit_ratio = random.randrange(10, 40) / 100
  profit.append( int(s * profit_ratio) )

employee = [int(p / 100000) for p in profit ]
```

```python
rawdata = pd.concat([pd.Series(careers, name='careers'), pd.Series(years, name='years'), pd.Series(sails, name='sails'), pd.Series(profit, name='profit'), pd.Series(employee, name='employee')], axis=1)
data = pd.DataFrame(rawdata)
data
```
